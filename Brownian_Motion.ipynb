```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters
n_steps = 1000       # Number of steps
step_size = 1        # Size of each random step

# Generate random steps for x, y, z directions
steps_x = np.random.choice([-step_size, step_size], size=n_steps)
steps_y = np.random.choice([-step_size, step_size], size=n_steps)
steps_z = np.random.choice([-step_size, step_size], size=n_steps)

# Cumulative sum to get positions
x = np.cumsum(steps_x)
y = np.cumsum(steps_y)
z = np.cumsum(steps_z)

# Plotting the 3D Brownian motion path
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, lw=0.7)
ax.scatter(x[0], y[0], z[0], color='green', label='Start', s=50)
ax.scatter(x[-1], y[-1], z[-1], color='red', label='End', s=50)

ax.set_title('3D Brownian Motion Simulation')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.show()

# write code in the next Session
```
